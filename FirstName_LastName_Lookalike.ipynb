```python
import pandas as pd

# Load the Customers dataset using a corrected file path
customers = pd.read_csv(r"C:\Users\DELL\Downloads\Customers.csv")  # Use raw string or forward slashes

# Function to calculate similarity score (example using a simple metric)
def calculate_similarity(customer_a, customer_b):
    # Example: Similarity based on region (you can enhance this)
    if customer_a['Region'] == customer_b['Region']:
        return 1.0  # Perfect match
    else:
        return 0.0  # No match

# Initialize a dictionary to hold lookalikes
lookalikes = {}

# Loop through the first 20 customers
for index_a in range(20):
    cust_a = customers.iloc[index_a]
    cust_id_a = cust_a['CustomerID']
    
    # List to hold (CustomerID, Score) tuples
    similar_customers = []
    
    # Compare with other customers
    for index_b in range(len(customers)):
        if index_a != index_b:  # Avoid comparing the same customer
            cust_b = customers.iloc[index_b]
            score = calculate_similarity(cust_a, cust_b)
            if score > 0:  # Only consider positive scores
                similar_customers.append((cust_b['CustomerID'], score))
    
    # Sort by score and take the top 3
    similar_customers.sort(key=lambda x: x[1], reverse=True)
    top_3_similar = similar_customers[:3]
    
    # Store in the dictionary
    lookalikes[cust_id_a] = top_3_similar

# Create a DataFrame from the lookalikes dictionary
lookalike_data = []
for cust_id, similar in lookalikes.items():
    for sim_cust_id, score in similar:
        lookalike_data.append([cust_id, sim_cust_id, score])

lookalike_df = pd.DataFrame(lookalike_data, columns=['CustomerID', 'LookalikeID', 'SimilarityScore'])

# Save to CSV
lookalike_df.to_csv('Lookalike.csv', index=False)

print("Lookalike.csv has been created successfully.")
```
